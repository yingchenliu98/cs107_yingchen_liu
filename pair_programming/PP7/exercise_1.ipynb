Part B:

$J_p(x,y) = \begin{bmatrix} 2x & 2y \\ e^{(x+y)} & e^{(x+y)}\end{bmatrix} \cdot p$



$J_p = \begin{bmatrix} 2x & 2y \\ e^{(x+y)} & e^{(x+y)}\end{bmatrix}\cdot \begin{bmatrix}1 \\ -2\end{bmatrix} = \begin{bmatrix}2x-4y \\ -e^{(x+y)}\end{bmatrix}$ for $p = \begin{bmatrix}1 \\ -2\end{bmatrix}$



$J_p= \begin{bmatrix} 2x & 2y \\ e^{(x+y)} & e^{(x+y)}\end{bmatrix}\cdot \begin{bmatrix}1 \\ 1\end{bmatrix} = \begin{bmatrix}2x+2y \\ 2e^{(x+y)}\end{bmatrix}  $for $p = \begin{bmatrix}1 \\ 1\end{bmatrix}$
